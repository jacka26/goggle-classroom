```python
import matplotlib.pyplot as plt
import numpy as np

def plot_trigonometric_function(function):
    x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

    if function == "sin":
        y = np.sin(x)
    elif function == "cos":
        y = np.cos(x)
    elif function == 'tan':
        y = np.tan(x)
    else:
        print("Incorrect. Please enter 'sin', 'cos', or 'tan'.")
        return

    plt.plot(x, y)
    plt.ylim(-5, 5) #https://stackoverflow.com/questions/54505852/how-to-plot-tanx-with-pyplot-and-numpy Since the tangent function diverges to infinity, plotting the full y range makes us miss things which is why I changed the range from -5 to 5.
    plt.title(f"Graph of {function}(x)")
    plt.xlabel("x")
    plt.ylabel(f"{function}(x)")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    function = input("Enter the trigonometric function you want to plot (sin, cos, or tan): ")
    plot_trigonometric_function(function)

    #I took some of this from stackoverflow, just the part abuot the sin, cos, and tan and the part where you print the output. I also used this youtube video, https://www.youtube.com/watch?v=5KitSbm7e1Q, this helped me find an equation that would work.
```
